# **Training The Models With Sampled** **Data**

**SELECTING THE DESIRED** **MODELS**

1.LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_curve, auc
lr = LogisticRegression()
# training the Logistic Regression Model with Training Data
lr.fit(X_train_resampled, y_train_resampled)
#testing the Logistic Regression Model with Testing Data
y_prediction_lr = lr.predict(X_test_resampled)
precision_log, recall_log, _ = precision_recall_curve(y_test_resampled, y_prediction_lr)
auprc_log = auc(recall_log, precision_log)
print('AUPRC (LogisticRegression):', auprc_log)
print(classification_report(y_test_resampled, y_prediction_lr))
print("Confusion Matrix is :",confusion_matrix(y_test_resampled, y_prediction_lr))
print("F1_Score of LogisticRegression is :",f1_score(y_test_resampled,y_prediction_lr))
acc_lr=accuracy_score(y_prediction_lr,y_test_resampled)
print("Acuracy_Score of LogisticRegression is :",acc_lr)
```

2.XGBClassifier

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(n_jobs=-1)
# training the RandomForestClassifier Model with Training Data
xgb.fit(X_train_resampled, y_train_resampled)
#testing the RandomForestClassifier Model with Testing Data
y_prediction_xgb = xgb.predict(X_test_resampled)
precision_xgb, recall_xgb, _ = precision_recall_curve(y_test_resampled, y_prediction_xgb)
auprc_xgb = auc(recall_xgb, precision_xgb)
print('AUPRC (XGBClassifier):', auprc_xgb)
print(classification_report(y_test_resampled, y_prediction_xgb))
print("Confusion Matrix is :",confusion_matrix(y_test_resampled, y_prediction_xgb))
acc_xgb=accuracy_score(y_prediction_xgb,y_test_resampled)
print("Acuracy_Score of XGBClassifier is :",acc_xgb)
print("F1 Score of XGBClassifier is :", f1_score (y_test_resampled,y_prediction_xgb))
```

3.KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier

# Train and evaluate KNN classifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_resampled, y_train_resampled)
y_prediction_knn = knn.predict(X_test_resampled)
precision_knn, recall_knn, _ = precision_recall_curve(y_test_resampled, y_prediction_knn)
auprc_knn = auc(recall_knn, precision_knn)
print('AUPRC (KNN):', auprc_knn)
# Summary of the predictions made by the classifier
print(classification_report(y_test_resampled, y_prediction_knn))
print("Confusion Matrix is :",confusion_matrix(y_test_resampled, y_prediction_knn))
# Accuracy score
acc_knn=accuracy_score(y_prediction_knn,y_test_resampled)
print('Accuracy of KNeighborsClassifier is',acc_knn)
```

4.LinearDiscriminantAnalysis

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
# Linear Discriminant Analysis
lda = LinearDiscriminantAnalysis()
lda.fit(X_train_resampled, y_train_resampled)
y_prediction_lda = lda.predict(X_test_resampled)
precision_lda, recall_lda, _ = precision_recall_curve(y_test_resampled, y_prediction_lda)
auprc_lda = auc(recall_lda, precision_lda)
print('AUPRC (LinearDiscriminantAnalysis):', auprc_lda)
# Summary of the predictions made by the classifier
print(classification_report(y_test_resampled, y_prediction_lda))
print("Confusion Matrix is :",confusion_matrix(y_test_resampled, y_prediction_lda))
# Accuracy score
acc_lda=accuracy_score(y_prediction_lda,y_test_resampled)
print('Accuracy of LinearDiscriminantAnalysis is',acc_lda)
print("F1 Score of LinearDiscriminantAnalysis is :", f1_score (y_test_resampled,y_prediction_lda))
```

5.Super Vector Machine

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train_resampled, y_train_resampled)
y_prediction_svc = svc.predict(X_test_resampled)
precision_svc, recall_svc, _ = precision_recall_curve(y_test_resampled, y_prediction_svc)
auprc_svc = auc(recall_svc, precision_svc)
print('AUPRC (SVC):', auprc_svc)
# Summary of the predictions made by the classifier
print(classification_report(y_test_resampled, y_prediction_svc))
print(confusion_matrix(y_test_resampled, y_prediction_svc))
# Accuracy score
acc_svc=accuracy_score(y_prediction_svc,y_test_resampled)
print('Accuracy of Super Vector Machine is',acc_svc)
```

6.KMeans

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(X_train_resampled, y_train_resampled)
test_pred_km = kmeans.predict(X_test_resampled)
acc_km=accuracy_score(y_test_resampled,test_pred_km)
precision_km, recall_km, _ = precision_recall_curve(y_test_resampled, test_pred_km)
auprc_km = auc(recall_km, precision_km)
print('AUPRC (KMeans):', auprc_km)
# Summary of the predictions made by the classifier
print(classification_report(y_test_resampled, test_pred_km))
# Accuracy score
print('Accuracy of KMeans is',acc_km)
```

```python
from sklearn.ensemble import RandomForestClassifier
rand = RandomForestClassifier()
# training the RandomForestClassifier Model with Training Data
rand.fit(X_train_resampled, y_train_resampled)
#testing the RandomForestClassifier Model with Testing Data
y_prediction_rand = rand.predict(X_test_resampled)
precision_rand, recall_rand, _ = precision_recall_curve(y_test_resampled, y_prediction_rand)
auprc_rand = auc(recall_rand, precision_rand)
print('AUPRC (RandomForestClassifier):', auprc_rand)
print(classification_report(y_test_resampled, y_prediction_rand))
print("Confusion Matrix is :",confusion_matrix(y_test_resampled, y_prediction_rand))
print("F1_Score of RandomForestClassifier is :",f1_score(y_prediction_rand,y_test_resampled))
acc_rand=accuracy_score(y_prediction_rand,y_test_resampled)
print("Acuracy_Score of RandomForestClassifier is :",acc_rand)
```
